# E-Commerce Sales Data Analysis

Business Problem :- 

The company wants to analyze its sales data to understand customer purchasing behavior, identify top-performing products, and improve revenue through data-driven decisions.

Objective :-

* Analyze sales trends over time
* Identify top-selling products
* Understand regional performance
* Find high-value customers
* Generate actionable business insights

# Import Libraries:- 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Dataset:-

```python
data = pd.read_csv(r"C:\Users\tarun\OneDrive\Desktop\small_raw_data.txt")
data
```

# Data Cleaning:- 

```python
data['Order_Date'] = pd.to_datetime(data['Order_Date'], errors = "coerce")
data
```

```python
data['Region'] = data['Region'].str.strip().str.title()
data
```

```python
data['Product'] = data['Product'].str.strip().str.title()
```

```python
data
```

# Create New Column (Revenue) :- 

```python
data['Revenue'] = data['Quantity'] * data['Price']
data
```

# Visualization:- 

# Monthly Sales Analysis :- 

```python
data['Month'] = data['Order_Date'].dt.month
```

```python
data
```

```python
monthly_sales = data.groupby('Month')['Revenue'].sum()

monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend', fontsize = 20, fontweight = "bold", color = "blue")
plt.xlabel('Month', fontsize = 15, fontweight = "bold", color = "blue")
plt.ylabel('Revenue', fontsize = 15, fontweight = "bold", color = "blue")
plt.xticks(rotation = 20, color = "k")
plt.yticks(rotation = 10)
plt.grid(False)
plt.show()
```

# Top Products:- 

```python
top_products = data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)

top_products.head(5).plot(kind='bar')
plt.title('Top 5 Products by Revenue',fontsize=20, fontweight = "bold", color = "blue")
plt.xlabel('Product',fontsize=20, fontweight = "bold", color = "blue")
plt.ylabel('Revenue',fontsize=20, fontweight = "bold", color = "blue")
plt.xticks(rotation = 20, color = "k", fontsize =12)
plt.yticks(rotation = 10, fontsize =12)
plt.grid(False)
plt.show()
```

# Sales by Region :- 

```python
region_sales = data.groupby('Region')['Revenue'].sum()

region_sales.plot(kind='bar')
plt.title('Sales by Region',fontsize=20, fontweight = "bold", color = "blue")
plt.xlabel('Region', fontweight = "bold", color = "blue",fontsize=20)
plt.ylabel('Revenue', fontweight = "bold", color = "blue",fontsize=20)
plt.xticks(rotation = 20, color = "k", fontsize =12)
plt.yticks(rotation = 10, fontsize =12)
plt.show()
```

# Category Analysis :- 

```python
category_sales = data.groupby('Category')['Revenue'].sum()

category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Category',fontsize=20, fontweight = "bold", color = "blue")
plt.ylabel('')
plt.show()
```

Key Insights :-
    
- Electronics category generates the highest revenue
- North region performs best in sales
- Laptop is the top-selling product
- Sales peak observed in February
